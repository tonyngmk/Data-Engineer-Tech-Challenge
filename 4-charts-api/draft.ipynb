### 1. Country-cases

```python
%%time

import time

start = '2022-11-01'
end = '2022-11-25'
asia_list = ['singapore', 'malaysia', 'vietnam', 'thailand', 'indonesia', 'taiwan', 'korea-south', 'korea-north']
df_list = []

for country in asia_list:
    url = f'https://api.covid19api.com/country/{country}/status/confirmed?from={start}T00:00:00Z&to={end}T00:00:00Z'
    data = requests.get(url)
    d = json.loads(data.content)
    df = pd.DataFrame(d)
    df_list.append(df)
    print(f'Completed: {country}')
    time.sleep(2)
```

```python
df = pd.concat(df_list).groupby(['Country', 'Lat', 'Lon', 'Date']).sum()
df = df.reset_index()
df = df.rename({"Lat":"lat", "Lon":"lon"}, axis=1)
```
